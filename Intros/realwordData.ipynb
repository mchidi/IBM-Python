```python
%%sql
-- What is the highest Safety Score?

SELECT MAX(Safety_Score) MAX_SAFETY_SCORE
FROM chicago_pub_sch 
```

```python
%%sql
--Which schools have highest Safety Score?

SELECT name_of_school, Safety_Score
FROM chicago_pub_sch
WHERE Safety_Score = (SELECT MAX(Safety_Score) FROM chicago_pub_sch)
```

#### -- What are the top 10 schools with the highest "Average Student Attendance"?

```python
%%sql

SELECT name_of_school, Average_Student_Attendance
FROM chicago_pub_sch
ORDER BY 2 DESC
NULLS LAST
LIMIT 10
```

```python
%%sql
-- Retrieve the list of 5 Schools with the lowest Average Student Attendance -- sorted in ascending order based on attendance

SELECT name_of_school, Average_Student_Attendance
FROM chicago_pub_sch
ORDER BY 2 
NULLS LAST
LIMIT 5
```

#### -- Which Schools have Average Student Attendance lower than 70%?

```python
%%sql

SELECT name_of_school, Average_Student_Attendance
FROM chicago_pub_sch
WHERE Average_Student_Attendance < '70%'
```

### -- Which Schools have Average Student Attendance lower than 70%?

```python
%%sql

SELECT name_of_school, Average_Student_Attendance
FROM chicago_pub_sch
WHERE CAST(SUBSTR(Average_Student_Attendance, 0, LENGTH(Average_Student_Attendance)) AS DOUBLE) < 70.0
LIMIT 10
```

#### -- Which Schools have Average Student Attendance lower than 70%?

```python
%%sql

SELECT name_of_school, Average_Student_Attendance
FROM chicago_pub_sch
WHERE CAST(REPLACE(Average_Student_Attendance, '%', "") AS DOUBLE) < 70.0
LIMIT 10
```

#### Get the total College Enrollment for each Community Area

```python
%%sql

SELECT community_area_name, SUM(college_enrollment) AS TOTAL_ENROLLMENT
FROM chicago_pub_sch
GROUP BY community_area_name
```

#### Get the 5 Community Areas with the least total College Enrollment sorted in ascending order

```python
%%sql

SELECT community_area_name, SUM(college_enrollment) AS TOTAL_ENROLLMENT
FROM chicago_pub_sch
GROUP BY community_area_name
ORDER BY 2
LIMIT 5
```

#### List 5 schools with lowest safety score.

```python
%%sql

SELECT name_of_school, Safety_Score
FROM chicago_pub_sch
ORDER BY Safety_Score
NULLS LAST
LIMIT 5
```

```python
%%sql

SELECT name_of_school, Safety_Score
FROM chicago_pub_sch
WHERE Safety_Score != 'None'
ORDER BY Safety_Score
LIMIT 5
```

#### Get the hardship index for the community area which has College Enrollment of 4368

```python
%%sql

SELECT hardship_index
FROM census_data
WHERE LOWER(community_area_name) = (
  SELECT LOWER(community_area_name)
  FROM chicago_pub_sch
  WHERE college_enrollment = 4368)
```

#### Get the hardship index for the community area which has the highest value for College Enrollment

```python
%%sql

SELECT D.hardship_index, C.community_area_name
FROM chicago_pub_sch C
JOIN census_data D ON LOWER(C.community_area_name) = LOWER(D.community_area_name)
WHERE college_enrollment = (
  SELECT MAX(college_enrollment) FROM chicago_pub_sch
)
```
