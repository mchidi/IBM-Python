```python
df1 = df.replace('?', np.NaN)

df = df1.dropna(subset=['price'], axis=0)
```

## Laptop Dataset


```python
df_laptop = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-Coursera/laptop_pricing_dataset_base.csv", header=None)

df_laptop.head()
```

### Adding Headers

```python
# Add Headers
headers = ["Manufacturer", "Category", "Screen", "GPU", "OS", "CPU_core", "Screen_Size_inch", "CPU_frequency", "RAM_GB", "Storage_GB_SSD", "Weight_kg", "Price"]

df_laptop.columns = headers

df_laptop.head()
```

#### Replace '?' with 'NaN'

```python
df_laptop.replace('?', np.NaN, inplace=True)
```

### Data columns
column "Weight_kg" and "Screen_Size_inch" need to be converted to float

```python
df_laptop.dtypes
```

```python
df_laptop.head(3)
```

```python
df_laptop.describe(include='all')
```

```python
df_laptop.info()
```

```python
# Changing Data types
df_laptop['Weight_kg'] = df_laptop['Weight_kg'].astype(float)
df_laptop['Screen_Size_inch'] = df_laptop['Screen_Size_inch'].astype(float)

df_laptop.dtypes
```
